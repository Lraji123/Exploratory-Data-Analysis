```python
#seasons for the data--> slower in summner than winter
#sprint 3-5
#summer 6-8
#winter12-2
#fall 9-11
#split between 2 decimals
```

```python
df2['race_month']=df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2.head()
```

```python
df2['race_season']=df2['race_month'].apply(lambda x: "winter" if x>11 else "fall" if x>8 else "summer" if x>5 else "spring")
```

```python
df2.head()
```

```python
df2.groupby('race_season')['Athlete average speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
